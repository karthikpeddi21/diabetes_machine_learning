importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data loading and anaylsing

```python
data = pd.read_csv('../input/diabetes-data-set/diabetes.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data['Outcome'].value_counts()
```

0-> label means non diabetic
1-> label means person had diabetics

```python
data.groupby('Outcome').mean()
```

```python
x = data.drop(columns = 'Outcome', axis=1)
y = data['Outcome']
```

```python
print(x.head)
```

```python
print(y)
```

Now we will standarize the data


```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

the standard scaler contains the following parameters

StandardScaler(copy=true, with_mean=true, with_std=true)

```python
standarized_data = scaler.transform(x)
```

```python
print(standarized_data)
```

```python
x = standarized_data
y = data['Outcome']
```

```python
print(x)
```

```python
print(y)
```





*TRAIN_TEST_SPLIT*

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

**Training the model**

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train, y_train)
```

The parameters availaible in svc is 
SVC(C=1.0, break_ties=False, cache_size=200, class_weight=None, coef0=0.0,
    decision_function_shape='ovr', degree=3, gamma='scale', kernel='linear',
    max_iter=-1, probability=False, random_state=None, shrinking=True,
    tol=0.001, verbose=False)

**Predicting accuracy**

```python
x_train_prediction = classifier.predict(x_train)
```

```python
train_data_prediction = accuracy_score(x_train_prediction, y_train)
```

```python
print(train_data_prediction)
```

**Accuracy on testing**

```python
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print(test_data_accuracy)
```

Make a predictive System

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```

here we can see that there  are no null values
